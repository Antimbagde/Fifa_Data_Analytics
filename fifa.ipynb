```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv('players_20.csv')
```

```python
fifa['nationality'].value_counts()
```

```python
fifa['nationality'].value_counts()[0:10]
```

```python
top_5_nationalities = fifa['nationality'].value_counts()[0:5]
```

```python
fifa['nationality'].value_counts()[0:5].keys()
```

```python
plt.figure(figsize=(8,5))
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.title('Top 5 Nationalities in FIFA Dataset')
plt.bar(top_5_nationalities.index, top_5_nationalities.values,color='green')
plt.show()

```

```python
player_salary = fifa[['short_name','wage_eur']]
```

```python
#salary of player
player_salary.head()
```

```python
#  sorted salary in decreasing order
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.head()
```

```python
plt.figure(figsize=(8,5))
plt.xlabel(' Players')
plt.ylabel('Salary')
plt.title('Top 5 Salaries of Player')
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color='pink')
plt.show()
```
